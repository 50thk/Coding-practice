```python
'''
짝지어 제거하기
해당 문제를 while문으로 replace를 사용해 해결하려고 시도했지만, replace의 변경을 원하는 parameter로 직접 변수를 입력해주어야 동작하는 듯.. 
stack을 이용하는 방식을 사용
'''

def solution(s):
    stack = []
    for x in s:
      # stack이 비었으면 stack에 x를 추가하고 temp에 마지막 stack을 저장
        if not stack:
            stack.append(x)
            temp = x
      # temp와 x가 같다면 stack[-1]을 제거한 후 stack이 비어있지 않다면 stack[-1]을 temp에 저장
        elif temp == x:
            stack.pop()
            if stack:
                temp = stack[-1]
      # stack에 x를 저장 temp에 마지막 stack을 저장
        else:
            stack.append(x)
            temp = x
    # stack이 비어있으면 1, 비어있지 않으면 0
    return 1 if not stack else 0

# temp 없이 코딩
def solution2(s):
    stack = []
    for x in s:
        if not stack:
            stack.append(x)
        elif stack and stack[-1] == x:
            stack.pop()
        else:
            stack.append(x)

    return 1 if not stack else 0
```

```python
s = 'baabaa'
solution(s), solution2(s)
```
