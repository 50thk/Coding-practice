```python
'''
 
  while replace   , replace   parameter     .. 
stack   
'''

def solution(s):
    stack = []
    for x in s:
      # stack  stack x  temp  stack 
        if not stack:
            stack.append(x)
            temp = x
      # temp x  stack[-1]   stack   stack[-1] temp 
        elif temp == x:
            stack.pop()
            if stack:
                temp = stack[-1]
      # stack x  temp  stack 
        else:
            stack.append(x)
            temp = x
    # stack  1,   0
    return 1 if not stack else 0

# temp  
def solution2(s):
    stack = []
    for x in s:
        if not stack:
            stack.append(x)
        elif stack and stack[-1] == x:
            stack.pop()
        else:
            stack.append(x)

    return 1 if not stack else 0
```
