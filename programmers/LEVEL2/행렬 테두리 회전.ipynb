```python
'''
2021 Dev-Matching: 웹 백엔드 개발자 챌린지 - 행렬 테두리 회전하기
단순 구현으로 완전탐색
시계방향으로 right, down, left, up의 순으로 회전을 구현
lower를 단순 정수비교가 아닌 요소들의 stack을 통해 최솟값만을 추출해 answer에 저장시키는 방법도 있음!
'''

def solution(rows, columns, queries):
    answer = []
    array = []
    for i in range(rows):
        t = []
        for j in range(columns):
            t += [j+1+i*columns]
        array.append(t)
        
    for q in queries:
        x1,y1,x2,y2 = q[0]-1,q[1]-1,q[2]-1,q[3]-1
        lower = 9999999999
        temp = array[x1][y1]
        for r in range(y1,y2):
            if temp < lower:
                lower = temp
            array[x1][r+1], temp = temp, array[x1][r+1]

        for d in range(x1,x2):
            if temp < lower:
                lower = temp
            array[d+1][y2], temp = temp, array[d+1][y2]
        for l in range(y2,y1,-1):
            if temp < lower:
                lower = temp
            array[x2][l-1], temp = temp, array[x2][l-1]
        for u in range(x2,x1,-1):
            if temp < lower:
                lower = temp
            array[u-1][y1], temp = temp, array[u-1][y1]
        answer.append(lower)

    return answer
```

```python
rows, columns = 6, 6
queries = [[2,2,5,4],[3,3,6,6],[5,1,6,3]]
solution(rows,columns,queries)
```
