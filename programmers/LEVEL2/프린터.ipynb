```python
from collections import deque
'''
첫 풀이
queue 자료구조로 출력되는 순서를 answer에 저장하여 location의 위치를 찾아 출력하는 방식
'''
def solution(priorities, location):
    id = [(0, p) if i != location else (1, p) for i, p in enumerate(priorities)]
    queue = deque(id)
    answer = []
    while queue:
        v = queue.popleft()
        if not queue:
            answer.append(v)
        else:
            for i, q in enumerate(queue):
                if v[1] < q[1]:
                    queue.append(v)
                    break
                elif i == len(queue)-1:
                    answer.append(v)
    return answer.index(id[location])+1
'''
answer의 list를 만들지 않고 index함수 없이 answer를 1씩 증가하여 몇번째인지 바로 return해주는 방식
'''
def solution2(priorities, location):
    id = [(i, p) for i, p in enumerate(priorities)]
    queue = deque(id)
    answer = 0
    while queue:
        v = queue.popleft()
        if not queue:
            answer += 1
            return answer
        for i, q in enumerate(queue):
            if v[1] < q[1]:
                queue.append(v)
                break
            elif i == len(queue)-1:
                answer += 1
                if v[0] == location:
                    return answer
```

```python
priorities = [2, 1, 3, 2]
location = 2
solution(priorities, location), solution2(priorities, location)
```
