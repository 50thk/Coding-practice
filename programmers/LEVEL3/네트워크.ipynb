```python
'''
bfs   
'''

def dfs(x, pc, connect):
    # 
    if connect[x] == True:
        return 0
    
    # 
    pc[x][x] = 0
    connect[x] = True
    for i in range(len(pc)):
        if pc[x][i] == 1:
            pc[x][i] = 0
            dfs(i, pc, connect)
    return 1

def solution(n, computers):
    answer = 0
    connect = [False] * n
    for i in range(n):
        answer += dfs(i, computers, connect)
    return answer
```

```python
n, computers = 3, [[1, 1, 0], [1, 1, 0], [0, 0, 1]]
solution(n, computers)
```
