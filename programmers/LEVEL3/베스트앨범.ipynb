```python
def solution(genres, plays):
    
    genres_total_plays = {}
    for g, p, _ in zip(genres,plays,range(len(genres))):
        
        if g not in genres_total_plays:
            genres_total_plays[g] = p
        else:
            genres_total_plays[g] += p
            
    genres_total_plays = dict(sorted(genres_total_plays.items(), key=lambda x: x[1], reverse=True))
    genres_kinds = list(genres_total_plays.keys())
    genres_plays_list = [[x,y,z] for x,y,z in zip(genres,plays,range(len(genres)))]
    
    answer = []
    for g in genres_kinds:
        genre = list(sorted(list(filter(lambda x: x[0] == g, genres_plays_list)), key=lambda x: x[1], reverse=True))
        if len(genre) == 1:
            ans = [genre[0][2]]
            answer.extend(ans)
        else:
            ans = [genre[0][2], genre[1][2]]
            answer.extend(ans)
    
    return answer
```
