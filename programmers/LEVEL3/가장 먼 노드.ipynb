```python
'''
  
 dfs/bfs      ...
 1       
1.    bfs   
2.  0 ,     1     1   
'''
from collections import deque
def solution(n, edge):
    def bfs(graph, visit, s):
      #   queue 
        visit[s] = 1
        queue = deque([s])
      # queue  
        while queue:
          # queue     pop
            v = queue.popleft()
          # v   for  visit[i] 0  queue   visit[i] visit[v]+1    
            for i in graph[v]:
                if not visit[i]:
                    queue.append(i)
                    visit[i] = visit[v]+1
        #       max      count 
        return visit.count(max(visit))
    #  
    visit = [0]*(n+1)
    #     dict 
    graph = {i:[] for i in range(n+1)}
    for a,b in edge:
        graph[a] += [b]
        graph[b] += [a]
    return bfs(graph, visit, 1)
```

```python
n = 6
vertex = [[3,6], [4,3], [3,2], [1,3], [1,2], [2,4], [5,2]]
solution(n,vertex)
```
