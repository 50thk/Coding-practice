```python
'''
가장 먼 노드
유튜브에서 dfs/bfs 설명 영상을 보고 처음으로 스스로 풀어보는...
시작점 1에서 가장 멀리 떨어져있는 노드의 개수를 구하는 문제
1. 넓이 탐색을 하는 bfs가 적절하다고 우선 판단
2. 노드가 0일 때, 방문을 하지 않은 것으로 1에서 하나씩 노드를 전진하며 거리를 1씩 늘려주는 방법으로 풀이
'''
from collections import deque
def solution(n, edge):
    def bfs(graph, visit, s):
      # 시작점을 방문하고 queue로 저장
        visit[s] = 1
        queue = deque([s])
      # queue가 빌 때까지
        while queue:
          # queue 구조에서 가장먼저 들어간 번호를 pop
            v = queue.popleft()
          # v노드에 연결된 번호들로 for문을 돌려 visit[i]가 0일 때 queue에 추가 및 visit[i]를 visit[v]+1하여 시작점에서의 거리를 기재 
            for i in graph[v]:
                if not visit[i]:
                    queue.append(i)
                    visit[i] = visit[v]+1
        # 문제의 가장 먼 노드의 개수들이기 때문에 max로 가장 먼 거리를 취하고 이를 count로 세어줌
        return visit.count(max(visit))
    # 방문 여부
    visit = [0]*(n+1)
    # 각 노드의 연결된 번호를 dict으로 변환
    graph = {i:[] for i in range(n+1)}
    for a,b in edge:
        graph[a] += [b]
        graph[b] += [a]
    return bfs(graph, visit, 1)
```

```python
n = 6
vertex = [[3,6], [4,3], [3,2], [1,3], [1,2], [2,4], [5,2]]
solution(n,vertex)
```
