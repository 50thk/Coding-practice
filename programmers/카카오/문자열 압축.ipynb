```python
"""
 2020   
"""
def solution(s):
    answer = len(s)
    if answer == 1:
        return 1
    for x in range(1, int(len(s)/2 + 1)):
        ss = []
        for i in range(0, len(s)+1, x):
            if not s[i:i+x]:
                break
            ss.append(s[i:i+x])
        temp = 1
        sss = ""
        for idx, i in enumerate(ss):
            if idx == len(ss)-1:
                if temp == 1:
                    sss += i
                else:
                    sss += str(temp) + i
                break
            if i == ss[idx+1]:
                temp += 1
            else:
                if temp == 1:
                    sss += i
                else:
                    sss += str(temp) + i
                    temp = 1
        if answer > len(sss):
            answer = len(sss)
    return answer
```

```python
s = "aabbaccc"
solution(s)
```
