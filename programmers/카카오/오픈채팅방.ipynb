```python
'''
카카오 2019 블라인드 채용 코테
'''
def solution(record):
    id_nick = {i.split()[1] : i.split()[2] for i in record if len(i.split()) == 3}
    answer = []
    for i in record:
        if i.split()[0] == "Change":
            continue
        elif i.split()[0] == "Enter":
            nick = id_nick[i.split()[1]]
            answer.append(f"{nick}님이 들어왔습니다.")
        else:
            nick = id_nick[i.split()[1]]
            answer.append(f"{nick}님이 나갔습니다.")
    return answer
    
```

```python
record = ["Enter uid1234 Muzi", "Enter uid4567 Prodo", "Leave uid1234", "Enter uid1234 Prodo", "Change uid4567 Ryan"]
solution(record)
```
